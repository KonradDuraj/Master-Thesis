```python
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline
```

```python
main_dir = os.getcwd()
filepath = os.path.join(main_dir, 'rewards.txt')
```

```python
rewards = []

with open(filepath) as file:
    
    lines = file.readlines()
    
    for line in lines:
        line = line.replace('[', '').split(sep = ',')

        for val in line:
            
            if val != ' \n' and val != '\n':
                
                if val.endswith(']\n'):
                    val = val.replace(']', '')
                    
                rewards.append(float(val))
file.close()
```

```python
plt.figure(figsize=(18,10))
plt.style.use('dark_background')
plt.grid(True)


plt.ylabel('Reward', color='white', fontsize = 20)
plt.xlabel('Episode', color = 'white', fontsize = 20)
plt.title('Positive rewards', fontsize=20)
plt.plot(index2, positive_rewards, color = 'red', marker = 'o',markerfacecolor='b',markeredgecolor='g', linestyle='--')
```

```python
def linreg(X, Y):
    """
    return a,b in solution to y = ax + b such that root mean square distance between trend line and original points is minimized
    """
    N = len(X)
    Sx = Sy = Sxx = Syy = Sxy = 0.0
    for x, y in zip(X, Y):
        Sx = Sx + x
        Sy = Sy + y
        Sxx = Sxx + x*x
        Syy = Syy + y*y
        Sxy = Sxy + x*y
    det = Sxx * N - Sx * Sx
    return (Sxy * N - Sy * Sx)/det, (Sxx * Sy - Sx * Sxy)/det

a,b = linreg(range(len(positive_rewards)),positive_rewards)

extrapolatedtrendline=[a*index + b for index in index2]

plt.figure(figsize=(18,10))
plt.style.use('dark_background')
plt.grid(True)


plt.ylabel('Reward', color='white', fontsize = 20)
plt.xlabel('Episode', color = 'white', fontsize = 20)
plt.title('Positive rewards with trend line', fontsize=20)
plt.scatter(index2, positive_rewards, color = 'red', marker = 'o')
plt.plot(extrapolatedtrendline,color = 'b', linewidth = 7 )
```
